```python
import pandas as pd
import numpy as np

data = {
    'patient_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110,
                   111, 112, 113, 114, 115, 101, 107, 118, 119, 120],
    'age': ['25', '34', None, '45', '29', None, '38', '52', '27', '41',
            '33', 'unknown', '48', '26', '35', '25', '38', '31', None, '44'],
    'weight': ['70', '65', '80', None, '75', None, '68', '90', '72', '85',
               '78', None, '82', '69', 'N/A', '70', '68', '74', None, '88'],
    'blood_pressure': [120, 130, None, 140, 125, None, 135, None, 118, 145,
                      128, None, 138, 122, None, 120, 135, 126, None, 142],
    'medication': ['Aspirin', 'Metformin', 'Lisinopril', None, 'Aspirin',
                   'Metformin', 'Lisinopril', 'Aspirin', None, 'Metformin',
                   'Lisinopril', 'Aspirin', None, 'Metformin', 'Aspirin',
                   'Aspirin', 'Lisinopril', 'Metformin', 'Aspirin', None],
    'insurance_provider': ['Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None,
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', 'Blue Cross',
                          'Aetna', None, 'UnitedHealth', 'Blue Cross', 'Aetna',
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None]
}

df = pd.DataFrame(data)
print(df)
```

Task 1: Inspect the Data

```python
#To view column data types and non-null counts
print("\n\n Column Data types and non-null counts")
print(df.info())

#To count missing values per column
print("\n\n Count of Missing Values per column")
print(df.isnull().sum())

#Percentage of missing values
print("\n\nPercentage of Missing Values")
print(df.isnull().sum() / len(df) * 100)

#To count duplicate rows
print("\n\nCount of Duplicate Rows")
print(df.duplicated().sum())
```

Task 2: Data Type Conversion

```python
df_new=df.copy()
# Convert age column to numeric
df_new['age']=pd.to_numeric(df_new['age'], errors='coerce')
print("\n Datatype of the Column age post conversion:",df_new["age"].dtype)

# Convert weight column to numeric
df_new['weight']=pd.to_numeric(df_new['weight'],errors='coerce')
print("\n Datatype of the Column weight post conversion:",df_new['weight'].dtype)

#To check how many NEW missing values were created after conversion
print("\n New Missing Values count after conversion:")
print(df_new.isnull().sum())

#Convert insurance_provider to category type using astype('category')
df_new['insurance_provider']=df_new['insurance_provider'].astype('category')
print("\n Datatype of the column insurance provider after conversion:",df_new['insurance_provider'].dtype)

#Use df.dtypes to verify all conversions
print("\n Datatypes of all columns after conversion:")
print(df_new.dtypes)
```

Task 3: Handle Missing Values

```python
#Age: Fill with median using fillna(df['age'].median())

df_new['age']=df_new['age'].fillna(df_new['age'].median())


#weight: Fill with median using fillna(df['weight'].median())
df_new['weight']=df_new['weight'].fillna(df_new['weight'].median())


#blood_pressure: Fill with median using fillna(df['blood_pressure'].median())
df_new['blood_pressure']=df_new['blood_pressure'].fillna(df_new['blood_pressure'].median())


#medication: Fill with mode using fillna(df['medication'].mode()[0])
df_new['medication']=df_new['medication'].fillna(df_new['medication'].mode()[0])

#insurance_provider: Fill with constant value 'Unknown'
df_new['insurance_provider'] = df_new['insurance_provider'].cat.add_categories(['Unknown'])
df_new['insurance_provider'] = df_new['insurance_provider'].fillna('Unknown')

print(df_new.isnull().sum())
```

Task 4: Handle Duplicates

```python
# View duplicate rows using df[df.duplicated(keep=False)]
print("\n Duplicate Rows:")
print(df_new[df_new.duplicated(keep=False)])

#Identify duplicates based on patient_id only using df.duplicated(subset=['patient_id']
print("\n Duplicate Rows based on patient_id:")
print(df_new[df_new.duplicated(subset='patient_id')])

```

```python
#Remove duplicates keeping first occurrence: df.drop_duplicates(subset=['patient_id'], keep='first'
df_new=df_new.drop_duplicates(subset='patient_id',keep='first')

#Print shape before and after to show how many rows were removed
print("\n Shape before removing duplicates:",df.shape)
print("\n Shape after removing duplicates:",df_new.shape)
```

Task 5: Complete Workflow with Verification

```python
df_clean = df.copy()

print("---------------Report Before Cleaning----------------")

print("\nDatatypes before cleaning:",df_clean.dtypes)
print("\nMissing values before cleaning:",df_clean.isnull().sum())
print("\nDuplicate values before cleaning:",df_clean.duplicated().sum())
print("\nShape before cleaning:",df_clean.shape)

print("\n---------------Report After Cleaning----------------")
df_clean['age']= pd.to_numeric(df_clean['age'],errors="coerce")
df_clean['weight']= pd.to_numeric(df_clean['weight'],errors="coerce")
df_clean['insurance_provider']= df_clean['insurance_provider'].astype('category')

print("\nDatatypes after cleaning:",df_clean.dtypes)

df_clean['age']=df_clean['age'].fillna(df_clean['age'].median())
df_clean['weight']=df_clean['weight'].fillna(df_clean['weight'].median())
df_clean['blood_pressure']=df_clean['blood_pressure'].fillna(df_clean['blood_pressure'].median())
df_clean['medication']=df_clean['medication'].fillna(df_clean['medication'].mode()[0])
df_clean['insurance_provider']=df_clean['insurance_provider'].cat.add_categories(['Unknown'])
df_clean['insurance_provider']=df_clean['insurance_provider'].fillna('Unknown')

print("\nMissing values after cleaning:",df_clean.isnull().sum())

df_clean=df_clean.drop_duplicates(subset='patient_id',keep='first')
print("\nDuplicate values post cleaning:",df_clean.duplicated().sum())

print("\nShape after cleaning:",df_clean.shape)
```
